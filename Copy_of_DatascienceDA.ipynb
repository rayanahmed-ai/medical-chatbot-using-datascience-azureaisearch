```python
# ================================
# DATA LOADING
# ================================
import kagglehub
from kagglehub import KaggleDatasetAdapter
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset 1: Heart Failure
df1 = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "fedesoriano/heart-failure-prediction",
    "heart.csv"
)

# Dataset 2: Kidney Disease
df2 = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "akshayksingh/kidney-disease-dataset",
    "kidney_disease.csv"
)

# Dataset 3: Heart Disease (300 rows)
df3 = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "ayushmanyashaswi/heart-disease-prediction-300-row-data-yashaswi",
    "heart_disease_data.csv"
)

# ================================
# DATA CLEANING & STANDARDIZATION
# ================================

df1_clean = df1[[
    "Age", "Sex", "RestingBP", "Cholesterol", "FastingBS", "HeartDisease"
]].copy()
df1_clean["Source_Dataset"] = "Heart Failure"

df2_clean = df2[[
    "age", "bp", "su", "classification"
]].copy()
df2_clean.columns = ["Age", "RestingBP", "FastingBS", "HeartDisease"]
df2_clean["Sex"] = pd.NA
df2_clean["Cholesterol"] = pd.NA
df2_clean["Source_Dataset"] = "Kidney Disease"

df3_clean = df3[[
    "age", "sex", "trestbps", "chol", "fbs", "target"
]].copy()
df3_clean.columns = [
    "Age", "Sex", "RestingBP", "Cholesterol", "FastingBS", "HeartDisease"
]
df3_clean["Source_Dataset"] = "Heart Disease 300"

# ================================
# DATASET COMBINATION
# ================================
combined_df = pd.concat(
    [df1_clean, df2_clean, df3_clean],
    ignore_index=True
)

# ================================
# MISSING VALUE HANDLING
# ================================
combined_df["RestingBP"].fillna(combined_df["RestingBP"].median(), inplace=True)
combined_df["Cholesterol"].fillna(combined_df["Cholesterol"].median(), inplace=True)
combined_df["FastingBS"].fillna(combined_df["FastingBS"].mode()[0], inplace=True)
combined_df.drop_duplicates(inplace=True)

# ================================
# FEATURE ENGINEERING
# ================================

# Age Group (Ordinal)
combined_df["AgeGroup"] = pd.cut(
    combined_df["Age"],
    bins=[0, 30, 45, 60, 100],
    labels=["Young", "Middle-aged", "Senior", "Elderly"]
)

# Blood Pressure Risk (Categorical)
combined_df["BP_Risk"] = pd.cut(
    combined_df["RestingBP"],
    bins=[0, 120, 140, 300],
    labels=["Normal", "Pre-Hypertension", "Hypertension"]
)

# Cholesterol Risk (Categorical)
combined_df["Chol_Risk"] = pd.cut(
    combined_df["Cholesterol"],
    bins=[0, 200, 240, 600],
    labels=["Normal", "Borderline", "High"]
)

# High Risk Indicator (Binary)
combined_df["HighRisk"] = (
    (combined_df["RestingBP"] > 140) &
    (combined_df["Cholesterol"] > 240)
).astype(int)

# ================================
# EXPLORATORY DATA ANALYSIS (EDA)
# ================================

# Age Distribution
plt.hist(combined_df["Age"], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Disease Outcome by Dataset
sns.countplot(x="Source_Dataset", hue="HeartDisease", data=combined_df)
plt.title("Disease Outcome Across Datasets")
plt.show()

# Disease Distribution by Age Group
sns.countplot(x="AgeGroup", hue="HeartDisease", data=combined_df)
plt.title("Disease Distribution by Age Group")
plt.show()

# Disease Outcome by BP Risk
sns.countplot(x="BP_Risk", hue="HeartDisease", data=combined_df)
plt.title("Disease Outcome by Blood Pressure Risk")
plt.show()

# Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(
    combined_df[["Age", "RestingBP", "Cholesterol", "FastingBS"]].corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Heatmap")
plt.show()

# ================================
# SAVE FINAL DATASET
# ================================
combined_df.to_csv("combined_data_with_features.csv", index=False)

print("✅ Feature engineering and EDA completed successfully.")
```
